## Scatter Plots

Scatter plots of (x,y) point pairs are created with matplotlib's ```plt.scatter()``` function. The required positional arguemnts are two lists or arrays. The first positional argument specifies the x-vales of each point and the second positional argument specifies the y-value of each point. The general form of a ```plot.plot()``` function call is:

```python
plt.scatter(x-points, y-points)
```

In the example below, 100 random, but semi-focused, x and y values are created.   These points are then plotted on a scatter plot using ```plt.scatter()```. Note the number of x-values is the same as the number of y-values.

````python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline


x1 = 1.5 * np.random.randn(100) + 10
y1 = 1.5 * np.random.randn(100) + 10
x2 = 1.5 * np.random.randn(100) + 4
y2 = 1.5 * np.random.randn(100) + 4
x = np.append(x1,x2)
y = np.append(y1,y2)

plt.scatter(x,y)
plt.show()
````

Matplotlib scatter plots can be customized by including keyword arguments in the ```plt.plot()``` function call. Note the keyword arguments used in ```plt.scatter()``` are a little different from the keyword arguments in other matplotlib plot types.

| scatter plot feature | ```plt.scatter()``` keyword | Example |
| --- | --- | --- |
| marker size | ```s=``` | ```plt.scatter(x, y, s=10)``` |
| marker color | ```c=``` | ```plt.scatter(x, y, c=(122, 80, 4))``` |
| marker opacity | ```alpha=``` | ```plt.scatter(x, y, alpha=0.1)``` |

Each of these keyword arguments can be assigned a an individual value which will apply to the whol scatter plot. The ```plt.scatter()``` keyword arguments can also be assigned to lists or arrays. Supplying a list or array controls the properties of each individual marker in the scatter plot.


````python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

x1 = 1.5 * np.random.randn(100) + 10
y1 = 1.5 * np.random.randn(100) + 10
x2 = 1.5 * np.random.randn(100) + 4
y2 = 1.5 * np.random.randn(100) + 4
x = np.append(x1,x2)
y = np.append(y1,y2)
colors = np.random.rand(100*2)
area = np.pi * (8 * np.random.rand(100*2))**2

plt.scatter(x, y,
            s=area,
            c=colors,
            alpha=0.5)
plt.show()
````
